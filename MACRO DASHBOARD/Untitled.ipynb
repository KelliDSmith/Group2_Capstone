# Tableau Visualization Dashboard#1 Data

```python
file = "Resources/Total_BLS_States2.csv"
total_bls_states = pd.read_csv(file)
total_bls_states.head()
```

```python
total_bls_states.info()
```

```python
file2 = "Resources/world_country_and_usa_states_latitude_and_longitude_values.csv"
lats_longs = pd.read_csv(file2)
lats_longs.head()
```

```python
lats_longs.info()
```

```python
merge_df = pd.merge(total_bls_states, lats_longs, on="usa_state")
merge_df.head()
```

```python
reduced_merged_df = merge_df.drop(['country_code', 'latitude', 'longitude', 'country'], axis=1)
reduced_merged_df
```

```python
renamed_merged_df = reduced_merged_df.rename(columns = {'Year': 'year', 'Gender': 'gender', 'Number\nof female\nworkers\n(in\nthousands)':'number of female workers(in thousands)', 'Number\nof male\nworkers\n(in\nthousands)': 'number of male workers(in thousands)', 'Median\nweekly\nearnings (female)': 'median weekly earnings(female)', 'Median weekly earnings (male)':'median weekly earnings(male)', 'Standard\nerror\nof\nmedian (female)': 'standard error of median (female)', 'Standard\nerror\nof\nmedian (male)': 'standard error of median (male)'})
renamed_merged_df
```

```python
renamed_merged_df.info()
```

```python
renamed_merged_df.to_csv('2015thru2020_BLS_State2.csv', index=False)
```
